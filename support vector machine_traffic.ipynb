```python
import pandas as pd
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
# load traffic data
traffic = pd.read_csv('LE5D_D.txt', sep='\t')
#traffic.head(30)
```

```python
# Define the features and target variables
features = ['tt_ex','nsnow', 'ntemp']
target = 'tt_vf'
```

```python
# Split the data into training and testing sets
train_data = traffic.sample(frac=0.8, random_state=1)
test_data = traffic.drop(train_data.index)
#train_data
```

```python
# Train a support vector regression model on the training data
regressor = SVR(kernel='linear', C=1.0, epsilon=0.1)
regressor.fit(train_data[features], train_data[target])
```

```python
# Predict traffic volume for the test data
predictions = regressor.predict(test_data[features])
print(predictions)
```

```python
# Calculate the mean absolute error of the predictions
mae = abs(predictions - test_data[target]).mean()

print("Mean absolute error: ", mae)
```
